```python
#convert timestamp version of column ("date") to "timestamp"
cl['timestamp'] = pd.to_datetime(cl['date'].astype(str), format='%Y%m%d') 
cl.drop('date', inplace=True, axis=1)
cl_hourly = cl.set_index("timestamp").resample("H").ffill().reset_index().rename(
    {"timestamp": "timestamp"}, axis=1
)
```

```python
mergeHourly = pd.merge(cl_hourly, johnson, left_on = 'timestamp', right_on = 'timestamp')
mergeHourly
```

# Data Cleaning (HOURLY DATA)

```python
print(type(mergeHourly['timestamp'][0]))
print(mergeHourly)
mergeHourly.dtypes
```

```python
new_df = mergeHourly
```

```python
query = """SELECT * FROM new_df WHERE TIME(timestamp) BETWEEN TIME('05:00:00') AND TIME('19:00:00') AND mean_power = 0"""
zero_df = ps.sqldf(query, locals())
zero_df
```

```python
zero_df['timestamp'] = pd.to_datetime(zero_df['timestamp'])
zero_df['timestamp']
```

```python
query2 = """SELECT AVG(mean_power) FROM new_df GROUP BY TIME(timestamp)"""
answer_avg = ps.sqldf(query2, locals())
answer_avg
```

```python
answer_avg = answer_avg.rename(columns={'AVG(mean_power)': 'Average Mean Power'})
answer_avg.index.names = ['Timestamp']
answer_avg
```

```python
std_df = new_df.groupby([new_df['timestamp'].dt.time])['mean_power'].std()
std_df
```

```python
std_list = []
for i in std_df:
    if str(i) == 'nan':
        i = 0.0
    std_list.append(float(i/1000))
std_list
#gets rid of NaN values/go back 2 decimal places (even though there are not null values present).
```
